# EDA: beir_nq



```python
import polars as pl
import os
import json
```

## Load Datasets

```python
PATH_BEIR_NQ = "../data/beir_nq"
corpus_list = [json.loads(line) for line in open(os.path.join(PATH_BEIR_NQ, "corpus.jsonl"))]
queries_list = [json.loads(line) for line in open(os.path.join(PATH_BEIR_NQ, "queries.jsonl"))]
corpus = pl.DataFrame(corpus_list).with_row_index(name="idx").drop("metadata")
queries = pl.DataFrame(queries_list).with_row_index(name="idx").drop("metadata").rename({"text": "question"})
```

## Saving processed dataframes

```python
PROCESSED_PATH = "../data/beir_nq/processed"
os.makedirs(PROCESSED_PATH, exist_ok=True)
corpus.write_ipc(f"{PROCESSED_PATH}/corpus.feather", compression="zstd")
queries.write_ipc(f"{PROCESSED_PATH}/queries.feather", compression="zstd")
```
